#1. Análisis de facturación



## 1.1 **Ingreso Total por tienda**

```python
tiendas['Venta_Total'] = tiendas['Precio'] + tiendas['Costo de envío']
tiendas.sample(10)
```

```python
tiendas['Venta_Total'].sum()
```

```python
tiendas.groupby('numero_de_tienda')[['Venta_Total']].sum().sort_values(by='Venta_Total', ascending=False)
```

## 1.2 **Cuadro de barras del total de las ventas por cada tiendas, emnumeradas por los numeros(1= tienda, 2= tiendas,3= tienda3, 4= tienda4)**

```python
facturacion = tiendas.groupby('numero_de_tienda')[['Venta_Total']].sum()\
               .sort_values(by='Venta_Total', ascending=False)

# Gráfico de barras
ax = facturacion.plot.bar(color ='red', y='Venta_Total', legend=False, figsize=(10, 6))

# Mostrar los valores reales encima de cada barra
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():,.0f}",     # Formato con separador de miles
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom'
    )

plt.title('Ventas Totales por tienda')
plt.ylabel('Monto de Ventas')
plt.xlabel('Venta total por tienda')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas.groupby('numero_de_tienda')[['Calificación']].mean().round(2)
```

```python
calificaciones = tiendas.groupby('numero_de_tienda')[['Calificación']].mean()

plt.figure(figsize=(12, 6))

# Gráfico de barras
barras_calificacion = calificaciones.plot(kind='bar', legend=False, figsize=(12,6), color= 'red')

# Títulos
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Número de Tienda')
plt.ylabel('Calificación Promedio')

# Mostrar valores encima de cada barra
for p in barras_calificacion.patches:
    barras_calificacion.annotate(
        f"{p.get_height():.2f}",                           # Valor con 2 decimales
        (p.get_x() + p.get_width() / 2, p.get_height()),   # Posición
        ha='center', va='bottom'
    )

plt.tight_layout()
plt.show()

```

# 4. Productos más y menos vendidos

## 4.1 **Top mas vendidos y menos vendidos tienda #1**

#### **Ventas maximas tienda #1**

```python
top10_maximos = (
    tienda.groupby('Producto')[['Venta_Total']].max()
    .sort_values(by='Venta_Total', ascending=False).head(10))

top10_maximos
```

```python
plt.figure(figsize=(22,12))

top10_maximos.sort_values(by='Venta_Total',ascending=True).plot(kind='barh', color= 'red')# Gráfico de barras horizontal

plt.title('Top 10 Productos con Mayor Venta Total (Valor Máximo)')#titulos
plt.xlabel('Venta Total')
plt.ylabel('Producto')

for i, v in enumerate(top10_maximos.sort_values(by= 'Venta_Total')['Venta_Total']):# Mostrar valores al final de cada barra
    plt.text(v, i, f"{v:,.0f}", va='center')

plt.tight_layout()
plt.show()
```

#### **Ventas minimas tienda #1**

```python
top10_minimos = (
    tienda.groupby('Producto')[['Venta_Total']]
          .min()
          .sort_values(by='Venta_Total', ascending=True)
          .head(10)
)

top10_minimos
```

```python
plt.figure(figsize=(22,12))

# Gráfico de barras horizontal
top10_minimos.sort_values(by='Venta_Total', ).plot(kind='barh', color= 'red')

# Títulos
plt.title('Top 10 Productos con Mayor Venta Total (Valor Máximo)')
plt.xlabel('Venta Total')
plt.ylabel('Producto')

# Mostrar valores al final de cada barra
for i, v in enumerate(top10_minimos.sort_values(by= 'Venta_Total')['Venta_Total']):
    plt.text(v, i, f"{v:,.0f}", va='center')

plt.tight_layout()
plt.show()
```
